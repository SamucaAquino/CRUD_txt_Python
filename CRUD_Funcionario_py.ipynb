```python
import os


def inserir(nome_arquivo, nome, matr, setor, cargo):
    with open(nome_arquivo, "a") as arquivo:
        arquivo.write(f'{nome},{matr},{setor},{cargo}\n')
        print(f'\nFuncionário: {nome} - Matrícula: {matr} - Setor: {setor} - Cargo: {cargo}\nCadastrado com sucesso.')

def exibir(nome_arquivo):
    try:
        with open(nome_arquivo, "r") as arquivo:
            linhas = arquivo.readlines()
            i = 1
            print('\n==========================================')
            for linha in linhas:
                nome, matr, setor, cargo = linha.strip().split(',')
                print(f'{i}) Nome: {nome}  - Matrícula: {matr} - Setor: {setor} - Cargo: {cargo}')
                i += 1
            print('============================================')
    except FileNotFoundError:
        print('Arquivo não encontrado!')
            
def consultar_cadastro(nome_arquivo, dados):
    try:
        with open(nome_arquivo, 'r') as arquivo:
            linhas = arquivo.readlines()
            resultados = []
            for linha in linhas:
                if dados.lower() in linha.lower():
                    resultados.append(linha.strip())
            if len(resultados) > 0:
                print("\nFuncionário Encontrado")
                for linha in resultados:
                    funcionario = linha.split(",")
                    print(f'Funcionário: {funcionario[0]} - Matrícula: {funcionario[1]} - Setor: {funcionario[2]} - Cargo: {funcionario[3]}')
            else:
                print(f"A informação fornecida não foi encontrada!")
    except FileNotFoundError:
        print(f"Arquivo: '{nome_arquivo}' não encontrado!")
    except Exception as e:
        print(f'Ocorreu um erro: {str(e)}')

def excluir_cadastro(nome_arquivo, dados):
    cadastro_excluido = False
    linhas = []
    with open(nome_arquivo, 'r') as arquivo:
        linhas = arquivo.readlines()
    with open(nome_arquivo, 'w') as arquivo:
        for linha in linhas:
            if dados not in linha:
                arquivo.write(linha)
            else:
                cadastro_excluido = True
    if cadastro_excluido:
        print('\nFuncionário excluído com sucesso!')
    else:
        print('\nNenhum funcionário foi excluído!')

def excluir_todo_cadastro(nome_arquivo):
    try:
        with open(nome_arquivo, 'r') as arquivo:
            linhas = arquivo.readlines()
            if len(linhas) == 0:
                print('\nNão há qualquer registro de cadastro de funcionário para excluir.')
            else:
                exibir(nome_arquivo)
                arquivo.close()
                mensagem = input('\nTem certeza que deseja excluir todos os cadastros (S/N)?') 
                if mensagem.lower() in ('s', 'sim'):
                    os.remove(nome_arquivo)
                    print('\nOs cadastros dos funcionários foram excluídos.')
                    print('\nArquivo vazio!')
                elif mensagem.lower() in ('n', 'não'):
                    print('\nOs cadastros dos funcionários não foram excluídos!')
                else:
                    print("\nEntrada Inválida!\nDigite apenas 'sim' ou 'não'!")             
    except Exception as e:
        print(f'Erro: {e}')

def alterar_cadastro(nome_arquivo, info, campo, novos_dados):
    with open(nome_arquivo, 'r') as arquivo_original:
        linhas = arquivo_original.readlines()
    with open(nome_arquivo, 'w') as arquivo_modificado:
        for linha in linhas:
            v_nome, matr, setor, cargo = linha.strip().split(',')
            if info in linha:
                if 'nome' in campo:
                    linha = linha.replace(v_nome, novos_dados)
                elif 'matrícula' in campo:   
                    linha = linha.replace(matr, novos_dados)
                elif 'setor' in campo:   
                    linha = linha.replace(setor, novos_dados)
                elif 'cargo' in campo:   
                    linha = linha.replace(cargo, novos_dados)
                else:
                    print('Campo inexistente!')
                arquivo_modificado.write(linha)
                funcionario = linha.split(',')
                print(f'\nFuncionário: {funcionario[0]} - Matrícula: {funcionario[1]} - Setor: {funcionario[2]} - Cargo: {funcionario[3]}')                
            else:
                arquivo_modificado.write(linha)    

def main():
    nome_arquivo = input(f'Digite o nome do arquivo (.txt): ')         

    while True:
        os.system('cls')
        menu = """\n
    -------------------------------
    MENU PRINCIPAL
    -------------------------------
    1) Inserir funcionário
    2) Exibir todos os funcionários
    3) Consultar Aluno
    4) Excluir funcionarios
    5) Excluir todos os funcionarios
    6) Alterar os dados do funcionário    
    7) Sair
    -------------------------------
    """
        print(menu)
        try:
            opcao = int(input('Digite a opção: '))          
            if opcao == 7:
                print('\nFim do Programa!')
                break
            elif opcao == 1:
                nome = input('\nNome do funcionário: ')
                matr = input('\nMatrícula do funcionário: ')
                setor = input('\nSetor do funcionário: ')
                cargo = input('\nCargo do funcionário: ')
                inserir(nome_arquivo, nome, matr, setor, cargo)
            elif opcao == 2:
                exibir(nome_arquivo)
            elif opcao == 3:
                dados = input('\nDigite o nome, sobrenome ou matrícula do funcionário que deseja consultar: ')
                consultar_cadastro(nome_arquivo, dados)
            elif opcao == 4:
                exibir(nome_arquivo)
                dados = input('\nDigite nome ou sobrenome, matrícula, setor ou cargo do funcionário: ')
                excluir_cadastro(nome_arquivo, dados)
            elif opcao == 5:
                excluir_todo_cadastro(nome_arquivo)
            elif opcao ==6:
                exibir(nome_arquivo)
                info = input('Informe o nome completo ou matrícula do funcionário que deseja atualizar cadastro: ')
                campo = input('Informe o campo que deseja atualizar o cadastro do funcionário: nome, matrícula, setor, ou cargo: ')
                if 'nome' in campo.lower():
                    novos_dados = input(f'Informe o novo {campo} do funcionário: ')
                elif 'matricula' or 'matrícula' in campo.lower():
                    novos_dados = input(f'Informe a nova {campo} do funcionário: ')
                elif 'setor' in campo.lower():
                    novos_dados = input(f'Informe o novo {campo} do funcionário: ')
                elif 'cargo' in campo.lower():
                    novos_dados = input(f'Informe o novo {campo} do funcionário: ')
                else:
                    print('Campo não encontrado')
                alterar_cadastro(nome_arquivo, info, campo, novos_dados)
        except ValueError:
            print('Digite um número!')



if __name__ == "__main__":
    main()
```
